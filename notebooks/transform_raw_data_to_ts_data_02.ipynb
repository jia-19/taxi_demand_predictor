```python
import pandas as pd

rides = pd.read_parquet("../data/transformed/validated_rides_2024_02.parquet")

rides.head(15)
```

```python
rides["pickup_hour"] = rides['pickup_datetime'].dt.floor('H')

rides
```

```python
#Manipulating tables to group our data in bins per hour per location
agg_rides = rides.groupby(['pickup_hour','pickup_location_id']).size().reset_index()
agg_rides.rename(columns = {0 : "total_pickups"}, inplace=True)

agg_rides
```
